# Import The Libraries For being used in this analysis

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Read The csv file

```python
data=pd.read_csv("Myntra_Data.csv",error_bad_lines=False)
```

```python
data
```

# Delete Unnecessary data Columns


```python

data=data.drop(['uniq_id','crawl_timestamp','variant_sku','dominant_material','dominant_color','body','product_details','complete_the_look','size_fit','type','inventory','specifications'],axis=1)
data
```

# which gender has more clothes?

```python

data['ideal_for'].value_counts()
```


# Show the clothes for women whose price is less than or equal to 1000


```python
data[(data['ideal_for']=='Women') & (data['variant_price']<=1000)]
```

# Show the brand name and counts the clothes of there brand


```python
data['brand'].value_counts()
```

# Show the clothes for men whose size is M and show there product quality

```python
data[(data['ideal_for']=='Men')& (data['size']=='M')]['care_instructions']
```

# Show the clothes for men whose price is 549 and data should be in stock and show there link

```python

data[(data['ideal_for']=='Men')& (data['variant_price']==549)& (data['is_in_stock']=='In Stock')]['link']
```

# Show the data for men and the product should be in stocks and how many quantities are there

```python
data[(data['ideal_for']=='Men')& (data['is_in_stock']=='In Stock')]['product_type'].value_counts()
```

# Show the data for men and there product type should be straight kurta and show there images

```python

data[(data['ideal_for']=='Men') & (data['product_type']=='Straight Kurta')]['images']
```

# How many color are there and count them

```python
data['actual_color'].value_counts()
```

# Draw a lineplot for showing the genders having more clothes

```python
temp=data['ideal_for'].value_counts()
sns.lineplot(data=temp)
plt.ylabel('No. of clothes')
plt.xlabel('Gender')
```

# Draw the scatter plot between variant_price and variant_compare_at_price

```python
sns.regplot(x=data['variant_price'],y=data['variant_compare_at_price'])
```
